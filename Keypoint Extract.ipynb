```python
import os
import pandas as pd
df=pd.DataFrame()

rootdir ='data/rps'
indexrow=0
for subdir,dirs, files in os.walk(rootdir):
    for file in files:
        print(file)
        label=str(subdir).split("\\")[1]
        f=open("data/txt/"+file[:-3]+"txt","r")
        a=f.readlines()[0]
        df.loc[indexrow,"filename"]=f.name[9:-4]
        a=a[1:-1].replace(" ","").split(",")
        for i in range(0,len(a),3):
            columnnamex="x"+str(a[i])
            columnnamey="y"+str(a[i])
            df.loc[indexrow,columnnamex]=a[i+1]
            df.loc[indexrow,columnnamey]=a[i+2]
        df.loc[indexrow,"label"]=label
        f.close()
        indexrow+=1
df
```
